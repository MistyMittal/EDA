```python
visa=pd.read_csv(r'C:\Users\user\Documents\naresh it 2\Desktop\naresh it\data files\visadataset_file.csv')
mean=round((visa['prevailing_wage'].mean()),2)
std=round(visa['prevailing_wage'].std(),2)
nr=visa['prevailing_wage']-mean
visa['prevailing_wage_z']=nr/std
visa
```

**Normalization**
- read the data again
- step1 : take the wage column
- step 2 :calculate the min value prevailing wage 
- step 3 :calculate the max of prevailing wage 
- step 4 :calculator the numerator : visa['prevailing_wage']-min
- step 5 :denominator step3-step2
- step 6:  divide step 4/step5

 
**package name : minmaxScaler**
 

```python
visa=pd.read_csv(r'C:\Users\user\Documents\naresh it 2\Desktop\naresh it\data files\visadataset_file.csv')
minn=round((visa['prevailing_wage'].min()),2)
maxx=round(visa['prevailing_wage'].max(),2)
nr=visa['prevailing_wage']-minn
dr=maxx-minn
visa['prevailing_wage_z']=nr/dr
visa[['prevailing_wage','prevailing_wage_z']]
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
visa['prevailing_wage_max_min']=mms.fit_transform(visa[['prevailing_wage']])
visa[['prevailing_wage','prevailing_wage_z','prevailing_wage_max_min']]
```
