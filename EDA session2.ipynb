**create data by using dictionary**

```python
name=['prachi','misty','shraddha']
age=[18,19,20]
pd.DataFrame(zip(name,age),columns=['Name','Age'])
```

```python
dict1={'Name':['prachi','misty','shraddha']
      ,'Age':[18,19,20]}
dict1
```

```python
# pass into dataframe
pd.DataFrame(dict1)
# automatically keys become columns
# and value become row
#no need to provide clumns name
# no zip nothing requried
```

```python
dict2={'Name':'Avinash',
       'Age':24,
       'City':'hyd'}
pd.DataFrame(dict2)
# it will show error as dict dost nat have list 
# so now we have to provide index 
#if list is there no need to provide index
```
