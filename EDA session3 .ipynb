$take$

- take is used to extract the particular data
- it have one parameter called axis
- axis=0 represents rows
- axis= 1 represent columns
- by default  axis= 0 is there
- in below example we given 100 200 300 and we did not provide axis value
- it understood we are talking about rows

```python
visa.take([100,200,300])
```

```python
visa.take([2,3,4],axis=1)
```

```python
visa.take([100:105])# not working
```

**draw back**
- you cannot find row or column in a single syntax
- you can't find by slicing [start:end] like above

```python
visa.take([100,200,300]).take([2,3,4],axis=1)
# need to take two time take 
# this drawbacks can be again by iloc
```
