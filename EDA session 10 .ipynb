**Data Transformation**
- Generally data has 3 types of distribution  
    - Noraml distribution
    - poditive skewed
    - negative skewed
- All the maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens  because of many reasons, in that one reason is outliers
- once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normality
- we have some  transformation methods mentioned below
    - Log transformation
    - Exponential Transformation
    -  Reciprocal Transformation
    -  Square root  Transformation
    -  Power  Transformation

**Reciprocal Transformation**
- Reciprocal will fail when values are low

```python
x=2
1/x
```

```python
reci_data=np.reciprocal(exp_data)
plt.hist(reci_data,bins=20,label='Reciprocal data')
plt.show()
```

**square root transfoemation**

```python
sqrt_data=np.sqrt(exp_data)
plt.hist(sqrt_data,bins=20,label='square root _data')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
log_data=np.log(exp_data)
plt.subplot(2,2,2).hist(log_data,bins=20,label='Logrithem')
plt.legend()
reci_data=np.reciprocal(exp_data)
plt.subplot(2,2,3).hist(reci_data,bins=20,label='RECIPROCAL')
plt.legend()
sqrt_data=np.sqrt(exp_data)
plt.subplot(2,2,4).hist(sqrt_data,bins=20,label='Square_root_data')
plt.legend()
plt.show()
```

**power transformer**

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans=pt.fit_transform(exp_data1)
plt.hist(data_trans)

# why we reshape it
# because  pwer transform requried  each value in list 
```
